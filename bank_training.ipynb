```python
import pandas as pd
```

```python
path = "E:/Datasets/bank/bank-processed.csv"
```

```python
df = pd.read_csv(path, index_col=0)
df
```

```python
x_data = df.drop(columns='y')
x_data.to
```

```python
y_data = df.y
y_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data,
                                                    test_size=0.40, random_state=42)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense, Dropout, BatchNormalization
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.metrics import accuracy_score
import keras
```

```python
model = Sequential()
model.add(Dense(x_train.shape[1], activation = 'relu', input_dim = x_train.shape[1]))
model.add(BatchNormalization())


model.add(Dense(128, activation = 'relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Dense(64, activation = 'relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Dense(32, activation = 'relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Dense(1, activation = 'sigmoid'))
```

```python
optimizer = keras.optimizers.Adam(lr=0.0001)
model.compile(optimizer = optimizer, loss = 'binary_crossentropy')
```

```python
history = model.fit(x=x_train, y=y_train, batch_size = 32, epochs=100, 
                    validation_data=(x_test, y_test), verbose=1)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
evaluation_metrics=pd.DataFrame(model.history.history)
evaluation_metrics.plot(figsize=(10,5))
plt.title("Loss for both Training and Validation", size = 20)
```

```python
y_pred = model.predict(x_test)
```
